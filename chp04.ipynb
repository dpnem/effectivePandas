```python
series = {
    'index':[0,1,2,3],
    'data':[145,142,38,12],
    'name':'songs'
}
```

```python
def get(series, idx):
    value_idx = series['index'].index(idx)
    return series['data'][value_idx]

get(series, 1)
```

```python
print(series['name'])
```
